```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the data_set
iris = sns.load_dataset("iris")
iris.head()
```

```python
#Determining the number of columns and rows.
iris.shape
```

```python
iris.columns
```

```python
iris.info()
```

```python
iris.describe()
```

```python
#Plotting a scatter plot between sepal_length and petal_length
plt.scatter(x=iris["sepal_length"], y=iris["petal_length"], color="blue")
plt.xlabel("sepal_length")
plt.ylabel("petal_length")
plt.title("sepal_length vs petal_length")
plt.show()
```

```python
plt.scatter(x=iris["sepal_width"], y=iris["petal_width"], color="blue")
plt.xlabel("sepal_width")
plt.ylabel("petal_width")
plt.title("sepal_width vs petal_width")
plt.show()
```

```python
#Plotting the histogram for all four columns
viz = iris
viz.hist()
plt.show()
```

```python
#plotting the boxplot for determining the outliers
sns.boxplot(iris)
plt.show()
```
